```python
list_dict = [dict1, dict2, dict3, dict4, dict5, dict6, dict7]
list_dict2 = [dict4, dict5, dict6, dict7]

mean_dict = {}
for key in dict1.keys():
    # mean_dict[key] = np.square(dict1[key].cpu().numpy())
    mean_dict[key] = dict1[key].cpu().numpy()
    for check_dict in list_dict[1:]:
        # mean_dict[key] += np.square(check_dict[key].cpu().numpy())
        mean_dict[key] += check_dict[key].cpu().numpy()

for key in dict1.keys():
    mean_dict[key] = mean_dict[key]/len(list_dict)

std_dev_dict = {}
for key in dict1.keys():
    # std_dev_dict[key] = np.square(np.square(dict1[key].cpu().numpy()) - mean_dict[key])
    std_dev_dict[key] = np.square(dict1[key].cpu().numpy() - mean_dict[key])
    for check_dict in list_dict[1:]:
        # std_dev_dict[key] += np.square(np.square(check_dict[key].cpu().numpy()) - mean_dict[key])
        std_dev_dict[key] += np.square(check_dict[key].cpu().numpy() - mean_dict[key])

for key in dict1.keys():
    std_dev_dict[key] = np.sqrt(std_dev_dict[key]/len(list_dict))
```

```python
new_list_dict = []
for check_dict in list_dict:
    new_check_dict = {}
    for key in check_dict.keys():
        new_check_dict[key] = check_dict[key].cpu().numpy().flatten()
    new_list_dict.append(new_check_dict)

max_dict = {}
min_dict = {}
for key in dict1.keys():
    temp_list = []
    for check_dict in new_list_dict:   
        temp_list.append(check_dict[key])
    out_max = np.max(np.array(temp_list), axis=0)
    out_min = np.min(np.array(temp_list), axis=0)
    out_max = out_max.reshape(dict1[key].shape)
    out_min = out_min.reshape(dict1[key].shape)
    max_dict[key] = out_max
    min_dict[key] = out_min
```
